```python
class FaceDataset(torch.utils.data.Dataset):
    def __init__(self, root_dir, transforms, mode='train'):
        self.root_dir= root_dir
        self.transforms= transforms

        self.img_files= sorted(glob('{}\\*.jpg'.format(root_dir)))
        self.json_files= sorted(glob('{}\\*.json'.format(root_dir)))
        # use 80% for training and 20% for validation... test set must be different
        if mode=='train':
            self.img_files= self.img_files[0:2143]
            self.json_files= self.json_files[0:2143]
        elif mode=='val':
            self.img_files= self.img_files[2143::]
            self.json_files= self.json_files[2143::]

    def read_json(self, file):
        with open(file) as f:
            data= json.load(f)
            data_string = json.dumps(data)
            test1 = Test(data_string)
            shapes = test1.shapes
            for i in range(len(shapes)):
                label.append(shapes[i]['label'])
                point = shapes[i]['points']
                point = np.array(point)
                points = point.flatten()
                coordinates.append(points)
        return label, coordinates

    def __getitem__(self, idx):
        img= Image.open(self.img_files[idx])
        label, points= self.read_json(self.json_files[idx])
        class_label= one_hot_encode_class_v2(label)
        bbox_label= torch.tensor(points, dtype= torch.float32)

        if transforms!=None:
            img= self.transforms(img)
        
        return (img, class_label, bbox_label)

    def __len__(self):
        return len(self.img_files)
```

```python
class_idx=['bodyfacemargo','bodyfaceagnes','bodyfacegru','bodyfaceedith','bodyfacebabycatpink','bodyfacebabycatpurple','bodyfacebabycatblue','bodyfacepinkcat']

def one_hot_encode_class_v2(label_list):
    labels=dict(zip(class_idx, [0,0,0,0,0,0,0,0]))
    
    for i in label_list:
        lower_case_string= i.lower()
        lower_case_string= lower_case_string.replace("_","")        
        for j in class_idx:
            if labels[j]!= 1:
                if lower_case_string == j:
                    labels[j]=1
                else:
                    labels[j]=0
        #print(i)
    return labels
# multi-label classification!
```
